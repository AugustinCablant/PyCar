```python
import pandas as pd
import os
import folium as fo
```

```python
cwd=os.getcwd()
accidents_2022_path=os.path.join(cwd,"carcteristiques-2022.csv")
accidents_2021_path= os.path.join(cwd,"carcteristiques-2021.csv")

accidents_2022_carac=pd.read_csv(accidents_2022_path,sep=";")
```

```python
accidents_2022_carac["lat"]=accidents_2022_carac["lat"].str.replace(',','.').astype(float)
accidents_2022_carac["long"]=accidents_2022_carac["long"].str.replace(',','.').astype(float)
```

```python
accidents_2022_carac.head()
```

```python
carte1=fo.Map([48.866667,2.333333],zoom_start=7)

#affichage de 1000 accidents
for i in range(1000):
    fo.Marker([accidents_2022_carac.iloc[i]["lat"],accidents_2022_carac.iloc[i]["long"]]).add_to(carte1)
    
carte1.save("test1.html")
```

```python
from folium.plugins import HeatMap

carte_chaleur=fo.Map([48.866667,2.333333],zoom_start=7)
#affichage de tous les accidents sous forme de carte de chaleur
list_coord=[]
for i in range(len(accidents_2022_carac)):
    list_coord.append([accidents_2022_carac.iloc[i]["lat"],accidents_2022_carac.iloc[i]["long"]])
    
HeatMap(list_coord).add_to(carte_chaleur)

carte_chaleur.save("carte_chaleur_accidents_france.html")
```
